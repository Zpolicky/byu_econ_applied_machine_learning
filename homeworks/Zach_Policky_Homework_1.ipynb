# Data set 1: CSV - Crime data US cities

```python
names = ['year_month', 'murder', 'rape', 'robbery', 'assault', 'burglary', 'larceny', 'vehicle_theft', 'population', 'city', 'ave_temp', 'prcp', 'snow']
crime = pd.read_csv("C:/Users/zachp/Dropbox/AAEcon/213r/crime_data.csv", header = None, names = names)
print(crime.head())

murd = crime['murder']
rape = crime['rape']
rob = crime['robbery']
assl = crime['assault']
burg = crime['burglary']
larc = crime['larceny']
veht = crime['vehicle_theft']
pop = crime['population']
city = crime['city']

```

```python
#check for missing data
crime.isnull()
```

```python
#since there doesnt seem to be a systematic missing of data we can just drop the missing values
crime1 = crime.dropna()
#Make a dataframe with the types of crimes I want to analyze
crime_df = crime1[['murder','rape', 'robbery', 'larceny', 'burglary', 'vehicle_theft']] 
```

```python
#describe the data
print('Full Data Set')
crime1.describe().round(2)

```

```python
print('Crime_df')
crime_df.describe().round(2)
```

```python
#quick galnce at average values
pd.pivot_table(crime_df, index = [city], aggfunc = np.mean).round(2)
```

```python
#I want to see which city has the worst murder rate
city_murder = crime.groupby(['city'])['murder'].mean()
city_murder.plot(kind = 'bar')
plt.ylabel('Murders')
plt.title("Average Murder Rates")
```

```python
#also vehicle theft rate
veht = crime.groupby(['city'])['vehicle_theft'].mean()
veht.plot(kind = 'bar')
plt.title('Average City Vehicle Theft')
```

```python
#seeing the correlations between variables is a quick way to visualize the data
crime_df.corr()
```

```python
g = sns.PairGrid(crime_df)
g.map(plt.scatter);
```

Conclusions:

Detroit has the highest crime rates as shown in the bar graphs. Also by looking at the scatter plots and correlation coefficients we can see that all types of crime are positively correlated in this data set.

```python
#Real GDP
gdp = fred.get_series('GDPC1')
#Consumer Price Index as a measure of inflation
cpi = fred.get_series('CPIAUCNS')
#GDP deflator as an alternative measure of inflation
gdp_def = fred.get_series('GDPDEF')
#unemployment 
unemp = fred.get_series("UNRATENSA")
```

```python
#check for missing values
data_df.isnull()
```

```python
#It looks like each series has different periods of measurement and different starting dates
#We can drop the the missing rows since there isn't a systematic method for missing data

data_df2 = data_df.dropna()
```

```python
data_df2.describe().round(2)
```

```python
#plot each series to see trends

data_df2["GDP"].plot()
plt.title("GDP")
plt.xlabel("Years")
```

```python
data_df2['CPI'].plot()
plt.xlabel('Years')
plt.title('Consumer Price Index')
```

```python
data_df2['GDP_Def'].plot()
plt.title('GDP Deflator')
plt.xlabel('Years')
```

```python
data_df['Unemp'].plot()
plt.xlabel('Years')
plt.title('Unemployment')
```

```python
u = data_df2['Unemp'].pct_change()
u.plot()
plt.title('% Change in Unemployment')
plt.xlabel('Years')
```

```python
g = data_df2['GDP'].pct_change()
g.plot()
plt.title('% Change in GDP')
plt.xlabel('Years')
```

```python
i = data_df2['CPI'].pct_change()
d = data_df2['GDP_Def'].pct_change()

#create a data frame with all the variables %change

change = [g,u,i,d]
change_df = pd.concat(change, axis = 1)
change_df.head()
```

```python
correlation = sns.PairGrid(change_df)
correlation.map(plt.scatter);
```

```python
change_df.corr()
```

Correlation Coefficients and Scatter plots tell us that increases in Unemployment and Inflation tend to yeild negative GDP change. Also increases in inflation when measured by the CPI tend to decrease Unemployment. 

```python
print("Variance in GDP")
sns.violinplot(data = gdp.pct_change(), color = 'red')
```

```python
print('Variance in CPI')
sns.violinplot(data = cpi.pct_change())
```

```python
print('Variance in GDP Deflator')
sns.violinplot(data = gdp_def.pct_change(), color = 'green')
```

```python
print('Variance in Unemployment')
sns.violinplot(data = unemp.pct_change(), color = 'orange')
```

The Violin plots tell us that there is the largest variance in GDP and that all variables are single mode variables.

```python
#seeing very few missing variables and a large sample size we can drop the rows with missing variables. 
#then we can check for outlyers in the variables of interest. Wage, hours, education

wage = wages.dropna()
wage['wage'].plot(kind = 'box')
plt.title('Wages Box Plot')
#we can see that there are outlyers in wage
```
